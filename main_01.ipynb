```python
import pandas as pd 
df = pd.read_csv('Food_Delivery_Times.csv')
```

```python
df2 = df.copy()
df2['Weather']=df2['Weather'].fillna(df2['Weather'].mode()[0])
df2['Traffic_Level']=df2['Traffic_Level'].fillna(df2['Traffic_Level'].mode()[0])
df2['Time_of_Day']=df2['Time_of_Day'].fillna(df2['Time_of_Day'].mode()[0])
df2['Courier_Experience_yrs']=df2['Courier_Experience_yrs'].fillna(df2['Courier_Experience_yrs'].mean())
```

```python
df.describe()
df2.info()
df2.isna().sum()
df2['Weather'].value_counts()
df2['Traffic_Level'].value_counts()

df2.to_csv('cleaned_ffod_delivery_times.csv',index=False)
```
